```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
xj = np.arange(10) #datos a interpolar
yj = np.sin(xj)
```

```python
Sj = CubicSpline(xj, yj)
x = np.arange(-0.5,9.6,0.01)
```

```python
plt.plot(xj,yj,"o",label="Datos") #Datos para hacer los splines
plt.plot(x,Sj(x),label="Interpolacion S") #Datos interpolados
plt.legend(loc='lower left', ncol=2)
plt.grid()
```

```python
plt.plot(x,Sj(x,1),label="S'") #Curva primera derivada
plt.plot(x,Sj(x,2),label="S''") #'''''' Segunda ''''''
plt.plot(x,Sj(x,3),label="S'''") #'''''' Tercera ''''''
plt.xlim(-0.5,9.5)
plt.legend(loc='lower left', ncol=2)
plt.grid()
```
